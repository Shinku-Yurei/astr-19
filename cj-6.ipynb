```python
# Trigonometric Function Definitions
# In The following cells, we will define two core trigonometric functions: Sine and Cosine.
# These functions will utilize Pythons built-in `math` library to perform the calculations.
```

```python
import math

def get_sin(x):
    """Returns the sine of x (where x is in radians)."""
    return math.sin(x)
```

```python
### Cosine Function
# Next, we define a function to return the Cosine of a value. 
# Like the sine function, this expects the input to be in radians.
import math

def cosine(x):
    return math.cos(x)
```

```python
def get_cos(x):
    """Returns the cosine of x (where x is in radians)."""
    return math.cos(x)
```

```python
# Create the data range
num_entries = 1000
lower_bound = 0
upper_bound = 2

# Calculate step size
step = (upper_bound - lower_bound) / (num_entries - 1)

# Generate lists of values (tabulating)
x_values = [lower_bound + (i * step) for i in range(num_entries)]
sin_values = [get_sin(x) for x in x_values]
cos_values = [get_cos(x) for x in x_values]

print(f"Successfully tabulated {len(x_values)} entries.")
```

```python
# Header for our table
print(f"{'x':>10} | {'sin(x)':>10} | {'cos(x)':>10}")
print("-" * 36)

# Using a for loop to print the first 10 entries
for i in range(10):
    x = x_values[i]
    s = sin_values[i]
    c = cos_values[i]
    print(f"{x:10.5f} | {s:10.5f} | {c:10.5f}")
```
